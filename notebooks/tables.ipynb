```python
import pandas as pd
from mtga_viz import *
from archetype_definition import dictionaries
```

```python
# Loading the collected data
df = pd.read_csv('../data/metagame_challenge/challenge_test_file.csv')
df_clean= clean_raw_csv(df, user_deck_col='user_deck', oppo_deck_col='oppo_deck')
df_w_extra_columns = create_new_columns(
    df_clean, 
    time_col='message_time_stamp',
    interval_window=1,
    run_col='run_result', 
    source_cols=['user_deck', 'oppo_deck'], 
    keyword_dict=dictionaries, 
    new_cols=['user_deck_arch', 'oppo_deck_arch'], 
    def_tag='other')

df_w_extra_columns.head(40)
```

```python
# This table will show you, for each deck in user_deck (or each archetype in user_deck_arch), the total amount of games, the wr with its error and the confidence level based on the amount of games.
get_table_wr_error(df_w_extra_columns, result_col='result_vs_oppo', study_col='user_deck')
```

```python
# This table will show you Korae's shared table (w/o the win rate)
get_table_runs(df_w_extra_columns, user_deck_col='user_deck', run_col='run_result')
```

```python
# This simple table is a quick summary with all contributing players, total games, amount of runs, event duration and # of every type of run
get_table_event_summary(df_w_extra_columns)
```

```python
# This function can be a little bit strange without proper plotting, but it will basically tell you how the share of those N most used decks fluctuates every hour along the event duration. 
get_window_deck_table(df_w_extra_columns, top_n=12)
```
