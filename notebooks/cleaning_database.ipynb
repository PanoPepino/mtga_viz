```python
import pandas as pd
from mtga_viz import *
```

```python
# Loading the collected data
df = pd.read_csv('../data/metagame_challenge/challenge_test_file.csv')
```

```python
# Homogenising decks (i.e. Izzet -> UR, GB -> BG, etc. Lower case everything)
df_clean= clean_raw_csv(df, user_deck_col='user_deck', oppo_deck_col='oppo_deck')
df_clean.head(10)
```

```python
# This function will extract the list of most played. It uses user_decks to calculate the intersection with most seen decks in the oppo_deck column
most_played_decks, common_deck_counting = explore_most_played_decks(df_clean, 'user_deck', 'oppo_deck')
```

```python
# This will tell you the most repeated decks in both user and oppo columns
list_to_table(most_played_decks,3)
```

```python
# Alternatively, one can also explore all the unique deck names for user_deck with:
list_to_table(df_clean['user_deck'].unique(),3)
```

```python
# With previous information, one can establish categories to assign archetypes to each deck. Here one can define which decks belong to which archetypes. No need to match full names. It could be single words in the full string.

# Better to have these definitions in a .py file to easily call them from other notebooks

TEMPO = ['tempo', 'shadow']
AGGRO = ['energy', 'aggro', 'bonds', 'shredder', 'burn', 'prowess', 'humans', 'goblins', 'affinity']
BLUE_COMBO = ['ub reanimator', 'show and tell', 'sneak and show', 'mill', 'nadu']
FAST_COMBO = ['breach', 'sorin', 'muxus']
CONTROL = ['control', 'beans']
STOMPY = ['stompy', 'prison']
MIDRANGE = ['goyf', 'lands']

dictionaries = {'tempo': TEMPO,
                'aggro' : AGGRO, 
                'blue_combo': BLUE_COMBO, 
                'fast_combo':FAST_COMBO,
                'stompy': STOMPY,
                'control': CONTROL,
                'midrange': MIDRANGE}
```

```python
# With previous information, we can then add the archetype tags. One can also extract total amount of trophies and amount of runs
df_w_extra_columns, amount_trophies, run_counting = create_new_columns(
    df_clean, 'message_time_stamp', 
    run_col='run_result', 
    source_cols=['user_deck', 'oppo_deck'], 
    keyword_dict=dictionaries, 
    new_cols=['user_deck_arch', 'oppo_deck_arch'], 
    def_tag='other')
```
